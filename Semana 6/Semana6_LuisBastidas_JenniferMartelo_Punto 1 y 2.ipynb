```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy import integrate
from scipy.optimize import fsolve
```

# 0.1 Ecuaci√≥n diferencial no lineal

```python
N = 1000
t = np.linspace(0.,10.,N+1)
r0 = np.array([1.])
q = np.array([[0., 0.2, 0.4],[0.7, 0.9, 1.]])
colors=np.array([['k','r','c'],['m','b','g']])
fig, axs = plt.subplots(2,3,  figsize=([18,6]))
for i in range(2):
    for j in range(3):
        def System(t, r0):
            x= r0
            return x**(q[i][j])
        solution = solve_ivp( System, (t[0],t[-1]), r0, t_eval = t, method = 'RK45' )
        t1 = solution.t
        x= solution.y[0]
        axs[i][j].plot(t1,x,color=colors[i][j], label=f"q={q[i][j]}")
        axs[i][j].legend()
```

1. Encuentre los coeficientes del m√©todo de Adams-Bashforth de tres puntos. s=3


```python
import sympy as sym
t = sym.Symbol('t',Real=True)
```

```python
F1 = (t+1)*(t+2)/(2)
F2 = -(t)*(t+2)/(1)
F3 = t*(t+1)/(2)
```

```python
I1 = sym.integrate(F1, (t,0,1))
I2 = sym.integrate(F2, (t,0,1))
I3 = sym.integrate(F3, (t,0,1))
print(I1,I2,I3)
```
